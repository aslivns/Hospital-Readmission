```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

- I did some web scraping about the topic but I will not use them at my presentation

```python
url = "https://www.hcup-us.ahrq.gov/reports/statbriefs/sb153.jsp"
```

```python
# 3. download html with a get request
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# after several trials I found the table
soup.select("body > div.bodycontainer tr td ")
```

```python
# percent admitted
soup.select("body > div.bodycontainer tr td ")[11].get_text()
```

- patterns in the table is +5 for diagnosis and percentage values, so to range I added 5 >>> (8,154,5)

```python
diagnosis = []
percentage = []

# for loop to retrive the data
for i in range(8,154,5):
    diagnosis.append(soup.select("body > div.bodycontainer tr td ")[i].get_text())
    percentage.append(soup.select("body > div.bodycontainer tr td ")[i+3].get_text())
```

- All-cause 30-day readmissions ranked by the most frequently treated conditions* in U.S. hospitals, 2010

```python
readmissions = pd.DataFrame({"principal_diagnosis":diagnosis,"percentage":percentage})
readmissions.head()
# All-cause 30-day readmissions ranked by the most frequently treated conditions* in U.S. hospitals, 2010
```

```python
readmissions.to_csv('most_freq_treated.csv', index = False)
```
